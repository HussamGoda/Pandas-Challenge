Local Government Area Summary

- Calculate the total number of schools
- Calculate the total number of students
- Calculate the total budget
- Calculate the average maths score
- Calculate the average reading score
- Calculate the percentage of students with a passing maths score (50 or greater)
- Calculate the percentage of students with a passing reading score (50 or greater)
- Calculate the percentage of students who passed maths and reading (% Overall Passing)
- Create a dataframe to hold the above results
- Optional: give the displayed data cleaner formatting

```python
unique_schools_number = o_schools_info_df["school_name"].count()
total_number_of_students = o_schools_info_df["size"].sum()
total_budget = o_schools_info_df["budget"].sum()
ave_maths_score = o_schools_students_info_df["maths_score"].mean()
ave_reading_score = o_schools_students_info_df["reading_score"].mean()
passing_maths_df = o_schools_students_info_df.loc[(o_schools_students_info_df["maths_score"] >= 50)]
number_students_passing_maths = passing_maths_df ["maths_score"].count()
ratio_passing_maths = 100 * (number_students_passing_maths / total_number_of_students)
passing_reading_df = o_schools_students_info_df.loc[(o_schools_students_info_df["reading_score"] >= 50)]
number_students_passing_reading = passing_reading_df ["reading_score"].count()
ratio_passing_reading = 100 * (number_students_passing_reading / total_number_of_students)
passing_maths_and_reading_df = o_schools_students_info_df.loc[(o_schools_students_info_df["maths_score"] >= 50) & (o_schools_students_info_df["reading_score"] >= 50)]
number_students_passing_maths_and_reading = passing_maths_and_reading_df ["reading_score"].count()
ratio_passing_maths_reading = 100 * (number_students_passing_maths_and_reading / total_number_of_students)
#print (unique_schools_number)
#print (total_number_of_students)
#print (total_budget)
#print (ave_maths_score)
#print (ave_reading_score)
#print (ratio_passing_maths)
#print (ratio_passing_reading)
#print (ratio_passing_maths_reading)
input_data_summary = {"Total Schools": [unique_schools_number], "Total Students": [total_number_of_students], "Total Budget":[total_budget], "Average Maths Score":[ave_maths_score],\
             "Average Reading Score": [ave_reading_score], "% Passing Maths": [ratio_passing_maths], "% Passing Reading": [ratio_passing_reading], "% Overall Passing": [ratio_passing_maths_reading]}
high_level_summary_df = pd.DataFrame(input_data_summary, columns=["Total Schools", "Total Students", "Total Budget", "Average Maths Score", "Average Reading Score", "% Passing Maths", "% Passing Reading", "% Overall Passing"])
high_level_summary_df["Total Students"] = high_level_summary_df["Total Students"].map("{:,}".format)
high_level_summary_df["Total Budget"] = high_level_summary_df["Total Budget"].map("${:,.2f}".format)
high_level_summary_df.head()
```
